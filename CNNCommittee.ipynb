```python
import os
import re
import random
import numpy as np
import pandas as pd
from six.moves import cPickle as pickle
from sklearn.metrics import classification_report, confusion_matrix
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.models import load_model

%matplotlib inline
```

```python
def getFileRoutes(folder_routes):
    file_routes = []

    for file_name in os.listdir(folder_routes):
        file_routes.append(folder_routes + file_name)

    return file_routes
```

```python
TRAINED_MODELS_FOLDER_ROUTE = '/content/drive/My Drive/Colab Notebooks/Deceptive Opinion Spam CNN Committee/trained_model/models/'
TEST_DATA_FOLDER_ROUTE = '/content/drive/My Drive/Colab Notebooks/Deceptive Opinion Spam CNN Committee/trained_model/test_data/'
TOKENIZER_FOLDER_ROUTE = '/content/drive/My Drive/Colab Notebooks/Deceptive Opinion Spam CNN Committee/trained_model/tokenizer/'
```

```python
trained_models_file_routes = getFileRoutes(TRAINED_MODELS_FOLDER_ROUTE)
test_data_file_routes = getFileRoutes(TEST_DATA_FOLDER_ROUTE)
tokenizer_file_routes = getFileRoutes(TOKENIZER_FOLDER_ROUTE)
```

```python
models_of_30_cnn = []

for trained_models_file_route in trained_models_file_routes:
    print('Loading ' + trained_models_file_route + ' model')
    models_of_30_cnn.append(load_model(trained_models_file_route))

print('Finished loading all models.')
```

```python
from random import sample

models_of_10_cnn = []

random_indexes_for_fastext_model = sample(range(0, 15), 5)
random_indexes_for_glove_model = sample(range(15, 30), 5)

for random_index in random_indexes_for_fastext_model:
    print('Loading ' + trained_models_file_routes[random_index] + ' model')
    models_of_10_cnn.append(load_model(trained_models_file_routes[random_index]))

for random_index in random_indexes_for_glove_model:
    print('Loading ' + trained_models_file_routes[random_index] + ' model')
    models_of_10_cnn.append(load_model(trained_models_file_routes[random_index]))

print('Finished loading all models.')
```

```python
for test_data_file_route in test_data_file_routes:
    with open(test_data_file_route, 'rb') as pickle_file :
            save = pickle.load(pickle_file)
            X_test = save['X_test']
            y_test = save['y_test']

    del save
```

```python
for tokenizer_file_route in tokenizer_file_routes:
    with open(tokenizer_file_route, 'rb') as pickle_file :
            save = pickle.load(pickle_file)
            tokenizer = save['tokenizer']

    del save
```

```python
predictions_of_committee_with_30_cnn = []

for model in models_of_30_cnn:
    print('Predicting using ' + str(model))
    predictions_of_committee_with_30_cnn.append(model.predict(X_test))

predictions_of_committee_with_30_cnn = np.asarray(predictions_of_committee_with_30_cnn)
cnn_committee_with_30_cnn_final_prediction = np.average(predictions_of_committee_with_30_cnn, axis = 0)
print('Finished predicting the data.')
```

```python
predictions_of_committee_with_10_cnn = []

for model in models_of_10_cnn:
    print('Predicting using ' + str(model))
    predictions_of_committee_with_10_cnn.append(model.predict(X_test))

predictions_of_committee_with_10_cnn = np.asarray(predictions_of_committee_with_10_cnn)
cnn_committee_with_10_cnn_final_prediction = np.average(predictions_of_committee_with_10_cnn, axis = 0)
print('Finished predicting the data.')
```

```python
random_fasttext_model_index = random.choice(range(0, 15))
print('Predicting using ' + str(models_of_30_cnn[random_fasttext_model_index]))
single_cnn_with_fasttext_final_prediction = models_of_30_cnn[random_fasttext_model_index].predict(X_test)
print('Finished predicting the data.')
```

```python
random_glove_model_index = random.choice(range(15, 30))
print('Predicting using ' + str(models_of_30_cnn[random_glove_model_index]))
single_cnn_with_glove_final_prediction = models_of_30_cnn[random_glove_model_index].predict(X_test)
print('Finished predicting the data.')
```

```python
print('==========================================================================')
print('============== Evaluation Report For 30 CNN Committee Model: =============')
print('==========================================================================')
print(classification_report(np.argmax(y_test, axis = 1), np.argmax(cnn_committee_with_30_cnn_final_prediction, axis = 1)))

print('==========================================================================')
print('============== Evaluation Report For 10 CNN Committee Model: =============')
print('==========================================================================')
print(classification_report(np.argmax(y_test, axis = 1), np.argmax(cnn_committee_with_10_cnn_final_prediction, axis = 1)))

print('===================================================================================')
print('============== Evaluation Report For Single CNN (FastText) Model: =================')
print('===================================================================================')
print(classification_report(np.argmax(y_test, axis = 1), np.argmax(single_cnn_with_fasttext_final_prediction, axis = 1)))

print('================================================================================')
print('============== Evaluation Report For Single CNN (GloVe) Model: =================')
print('================================================================================')
print(classification_report(np.argmax(y_test, axis = 1), np.argmax(single_cnn_with_glove_final_prediction, axis = 1)))
```
